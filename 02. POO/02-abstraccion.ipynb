## Abstracción:
El propósito es poder dotarnos de las herramientas conceptuales para poder modelar el mundo de acuerdo a la concepción personal y traducir esa concepción a código.
* Enfocarnos en la información relevante.
* Separar la información central de los detalles secundarios.
* Podemos usar variables y métodos (privados o públicos).

La abstracción lo podemos ver en todo lugar, por ejemplo: Cuando conducimos un vehiculo, no nos detenemos a pensar como funciona el motor de forma detallada, no nos interesa la forma en la que se quema la gasolina o se hace combustión, como trabaja la suspensión para evitar movimientos abruptos, etc. Quizas cuando estamos en un elevados, tan solo nos detenemos a pensar en presionar el boton para que nos deje en un piso determinado, pero no nos preocupamos en las poleas, o como funciona....
Las abstracciones las vemos día a día en el mundo real.

¿Cómo traducimos estos conceptos en programación?
A través de las interfaces que diseñamos.

```python
class Automovil:
    def __init__(self, marca, modelo, color) -> None:
        self.marca = marca
        self.modelo = modelo
        self.color = color
    
    def __str__(self) -> str:
        return f"Marca: {self.marca}, Modelo: {self.modelo}, Color: {self.color}"

if __name__ == "__main__":
    auto = Automovil("Toyota", 2023,"Negro")
    print(auto)
    
```

```python

class Lavadora:
    def __init__(self):
        pass
    
    def lavar(self, temperatura='caliente'):
        self._llenar_agua(temperatura)
        self._dosificar_detergente()
        self._lavar()
        self._centrifugar()
            
    def _llenar_agua(self, temperatura):
        print(f"Llenando agua {temperatura}")
        
    def _dosificar_detergente(self):
        print("Dosificando detergente.")
            
    def _lavar(self):
        print("Lavando la ropa.")
        
    def _centrifugar(self):
        print("Centrifugando la ropa.")
        
if __name__ == "__main__":
    lavadora = Lavadora()
    lavadora.lavar()
        
```

```python
# Definición de la clase Persona:
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")

# Creación de objetos
persona1 = Persona("Juan", 25)
persona2 = Persona("María", 30)

# Acceso a atributos y métodos
print(persona1.nombre)  # Imprime "Juan"
print(persona2.edad)    # Imprime 30
persona1.saludar()      # Imprime "Hola, soy Juan y tengo 25 años."
persona2.saludar()      # Imprime "Hola, soy María y tengo 30 años."
```

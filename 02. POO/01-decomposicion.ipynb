## Decomposición:

Cuando nos encontramos con clases que pueden tener muchas propiedades, que además las lineas de código que puedan generarse sean muy extensas, se debe decomponer y asociar a otras clases para separar los métodos.

```python
# Clase Automovil:
class Automovil:
    
    def __init__(self, modelo, marca, color) -> None:
        self.modelo = modelo
        self.marca = marca
        self.color = color
        self._estado = 'en reposo'
        self._motor = Motor(cilindrada=1300)
    
    def acelerar(self, tipo='despacio'):
        if tipo == 'rapida':
            self._motor.inyecta_gasolina(10)
        else:
            self._motor.inyecta_gasolina(3)
        self._estado = 'en movimiento'

# Calse Motor:    
class Motor:
    
    def __init__(self, cilindrada, tipo = 'gasolina') -> None:
        self.cilindrada = cilindrada
        self.tipo = tipo
        self._temperatura = 0

    def get_cilindrada(self):
        return self.cilindrada

    def set_cilindrada(self, value):
        self.cilindrada = value

    def get_tipo(self):
        return self.tipo

    def set_tipo(self, value):
        self.tipo = value

    @property
    def temperatura(self):
        return self._temperatura

    @temperatura.setter
    def temperatura(self, value):
        self._temperatura = value
        
    def inyecta_gasolina(self, cantidad):
        pass

if __name__ == "__main__":
    automovil1 = Automovil("2020","Toyota","Azul")
    automovil1.acelerar
    
        
```
